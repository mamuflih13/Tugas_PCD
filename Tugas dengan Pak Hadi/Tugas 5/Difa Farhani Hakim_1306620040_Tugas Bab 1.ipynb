```python
from PIL import Image as img
from PIL import ImageEnhance as imgEn
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

# Load Gambar

```python
image = img.open('Difa_gambar.jpg')
```

```python
plt.figure(figsize=(8,8))
plt.imshow(image)
plt.show()
```

# Transformasi

## Rotasi

```python
img_r45 = image.rotate(45)
img_r90 = image.rotate(90)
img_r180 = image.rotate(180)
img_r270 = image.rotate(270)
```

```python
plt.figure(figsize=(8,8))
plt.subplot(2,2,1)
plt.title('Rotate 45 Degree')
plt.imshow(img_r45)
plt.subplot(2,2,2)
plt.title('Rotate 90 Degree')
plt.imshow(img_r90)
plt.subplot(2,2,3)
plt.title('Rotate 180 Degree')
plt.imshow(img_r180)
plt.subplot(2,2,4)
plt.title('Rotate 270 Degree')
plt.imshow(img_r270)
plt.show()
```

## Affine

### Scaling

```python
width, height = image.size
size05 = (int(width/2), int(height/2))
size2 = (width*2, height*2)
size4 = (width*4, height*4)
img_scl05 = image.resize(size05)
img_scl2 = image.resize(size2)
img_scl4 = image.resize(size4)
```

```python
plt.figure(figsize=(8,8))
plt.subplot(2,2,1)
plt.title('Original Image')
plt.imshow(image)
plt.subplot(2,2,2)
plt.title('Scaling 0.5x')
plt.imshow(img_scl05)
plt.subplot(2,2,3)
plt.title('Scaling 2x')
plt.imshow(img_scl2)
plt.subplot(2,2,4)
plt.title('Scaling 4x')
plt.imshow(img_scl4)
plt.show()
```

### Translation

```python
img_trans_x100 = image.rotate(angle=0,translate=(100,0))
img_trans_y100 = image.rotate(angle=0,translate=(0,100))
img_trans_xy100 = image.rotate(angle=0,translate=(100,100))
```

```python
plt.figure(figsize=(8,8))
plt.subplot(2,2,1)
plt.title('Original Image')
plt.imshow(image)
plt.subplot(2,2,2)
plt.title('Translate x-axis 100px')
plt.imshow(img_trans_x100)
plt.subplot(2,2,3)
plt.title('Translate y-axis 100px')
plt.imshow(img_trans_y100)
plt.subplot(2,2,4)
plt.title('Translate x & y-axis 100 px')
plt.imshow(img_trans_xy100)
plt.show()
```

### Horizontal Shear

```python
img_horizontal_r = image.transform((int(1.4*image.width), image.height), img.AFFINE,
                    data=(1,-0.5,0,0,1,0))
img_horizontal_l = image.transform((int(1.4*image.width), image.height), img.AFFINE,
                    data=(1,0.5,-350,0,1,0))
```

```python
plt.figure(figsize=(8,8))
plt.subplot(2,2,1)
plt.title('Shear Right Bottom')
plt.imshow(img_horizontal_r)
plt.subplot(2,2,2)
plt.title('Shear Left Bottom')
plt.imshow(img_horizontal_l)
plt.show()
```

### Vertical Shear

```python
img_vertical = image.transform((int(1.4*image.width), image.height), img.AFFINE,
                    data=(1,0,0,0.5,1,0))
```

```python
plt.figure(figsize=(8,8))
plt.title('Vertical')
plt.imshow(img_vertical)
plt.show()
```

## Perspektif

```python
params = [1, 0.1, 0, -0.1, 0.5, 0, -0.005, -0.001]
img_perspective = image.transform((image.width//3, image.height), img.PERSPECTIVE, params,
img.BICUBIC)
```

```python
plt.figure(figsize=(8,8))
plt.title('Perspective')
plt.imshow(img_perspective)
plt.show()
```

## Pencerminan

```python
img_flip_LR = image.transpose(img.FLIP_LEFT_RIGHT)
```

```python
plt.figure(figsize=(8,8))
plt.subplot(2,2,1)
plt.title('Original')
plt.imshow(image)
plt.subplot(2,2,2)
plt.title('Mirroring Left to Right')
plt.imshow(img_flip_LR)
plt.show()
```

# Plot RGB Channel

```python
ch_r, ch_g, ch_b = image.split()
```

```python
plt.figure(figsize=(18,6))
plt.subplot(1,3,1)
plt.imshow(ch_r, cmap=plt.cm.Reds)
plt.subplot(1,3,2)
plt.imshow(ch_g, cmap=plt.cm.Greens)
plt.subplot(1,3,3)
plt.imshow(ch_b, cmap=plt.cm.Blues)
plt.tight_layout()
plt.show()
```

# Plot Histogram RGB
