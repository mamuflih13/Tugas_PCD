```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image, ImageOps

#Showing Image
image = Image.open('jerapah.jpg')
plt.imshow(image)
```

**Transformasi rotasi**

```python
Image_45 = image.rotate(45) # rotate the image by 45 degrees
plt.imshow(Image_45) # show the rotated image
```

```python
Image_90 = image.rotate(90) # rotate the image by 45 degrees
plt.imshow(Image_90) # show the rotated image
```

```python
Image_180 = image.rotate(180) # rotate the image by 45 degrees
plt.imshow(Image_180) # show the rotated image
```

**Transformasi Affine**

```python
affine = image.transform((int(1.4*image.width), image.height), Image.Transform.AFFINE, data=(1,-0.5,0,0,1,0))
plt.imshow(affine)
```

**Transformasi Perspektif**

```python
params = [1, 0.1, 0, -0.1, 0.5, 0, -0.005, -0.001]
perspective = image.transform((image.width//3, image.height), Image.Transform.PERSPECTIVE, params, Image.Resampling.BICUBIC)
plt.imshow(perspective)
```

**Pencerminan**

```python
im_mirror = ImageOps.mirror(image)
plt.imshow(im_mirror)
```

**Plot Gambar Masing-Masing Nilai RGB Secara Terpisah**

```python
#Separating the RGB Channels of an image
ch_r, ch_g, ch_b = Image.split()
plt.figure(figsize=(18,6))
plt.subplot(1,3,1); plt.imshow(ch_r, cmap=plt.cm.Reds); plt.axis('off')
plt.subplot(1,3,2); plt.imshow(ch_g, cmap=plt.cm.Greens); plt.axis('off')
plt.subplot(1,3,3); plt.imshow(ch_b, cmap=plt.cm.Blues); plt.axis('off')
plt.tight_layout()
plt.show()
```

**Plot Histogram RGB Gambar**

```python
#Plotting the histograms of pixel values for the RGB channels of an image
pl = Image.histogram()
plt.bar(range(256), pl[:256], color='r', alpha=0.5)
plt.bar(range(256), pl[256:2*256], color='g', alpha=0.4)
plt.bar(range(256), pl[2*256:], color='b', alpha=0.3)
plt.show()
```
