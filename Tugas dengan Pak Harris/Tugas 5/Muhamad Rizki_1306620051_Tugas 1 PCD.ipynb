**PENGOLAHAN CITRA DIGITAL**

```python
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image = Image.open('Landscape.jpg')
plt.imshow(image)
```

**Rotasi Gambar**

```python
im_45 = image.rotate(45)
plt.imshow(im_45)
```

```python
image_rot_90 = image.rotate(90)
plt.imshow(image_rot_90)
```

```python
image_rot_180 = image.rotate(180)
plt.imshow(image_rot_180)
```

**Affine**

```python
affine = image.transform((int(1.4*image.width), image.height), Image.AFFINE, data=(1,-0.5,0,0,1,0))
plt.imshow(affine)
```

**Perspektif**

```python
params = [1, 0.1, 0, -0.1, 0.5, 0, -0.005, -0.001]
perspektif = image.transform((image.width//3, image.height), Image.PERSPECTIVE, params, Image.BICUBIC)
plt.imshow(perspektif)
```

**Nilai RGB Secara Terpisah**

```python
ch_r, ch_g, ch_b = image.split()
plt.figure(figsize=(18,6))
plt.subplot(1,3,1); plt.imshow(ch_r, cmap=plt.cm.Reds); plt.axis('off')
plt.subplot(1,3,2); plt.imshow(ch_g, cmap=plt.cm.Greens); plt.axis('off')
plt.subplot(1,3,3); plt.imshow(ch_b, cmap=plt.cm.Blues); plt.axis('off')
plt.tight_layout()
plt.show()
```

**Plot Histogram RGB**

```python
pl = image.histogram()
plt.bar(range(256), pl[:256], color='r', alpha=0.5)
plt.bar(range(256), pl[256:2*256], color='g', alpha=0.4)
plt.bar(range(256), pl[2*256:], color='b', alpha=0.3)
plt.show()
```
